# Image Data Processing and Visualization with TensorFlow

## 1- Import Libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
```

## 2- Dataset Retrieval 

```python
# Download and Extract the Dataset

# Specify the URL of the Flower Photos dataset:

url = 'https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz'

# Define the local directory:

dataset_dir = './flower_photos'

# Use get_file to download and extract the dataset:

flowers = tf.keras.utils.get_file(
    'flower_photos', origin=url, untar=True, cache_subdir=dataset_dir
)

# Print the path:

print("Dataset is downloaded and extracted at:", flowers)
```

## 3- Data Preprocessing

```python
# Download and extract the Flower Photos dataset:

flowers = tf.keras.utils.get_file(
    'flower_photos',
    'https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz',
    untar=True
)

# Create an ImageDataGenerator with rescaling and rotation augmentation:

img_gen = ImageDataGenerator(rescale=1./255, rotation_range=20, )
```

## 4- Dataset Creation

```python
# Create a dataset from the image data generator:

ds = tf.data.Dataset.from_generator(
    lambda: img_gen.flow_from_directory(flowers, batch_size=32),
    output_types=(tf.float32, tf.float32),
    output_shapes=([32, 256, 256, 3], [32, 5])
)
```

```python
i = ds.take(1).as_numpy_iterator()
```

### I create a dataset iterator 

```python
ds2 = ds.take(110)
```

### asserting that each batch has the correct shape

```python
for i in ds.take(110):
    assert i[0].shape[0] == 32
#     break
```

## 5- Dataset Exploration and Visualization

```python
# Display a batch of images using matplotlib:

def visualize_images(images_batch):
    plt.figure(figsize=(10, 10))
    for n in range(images_batch.shape[0]):
        plt.subplot(4, 8, n + 1)
        plt.imshow(images_batch[n])
        plt.axis('off')
    plt.show()

# Take a batch of images from the dataset:

images_batch, _ = next(iter(ds))
visualize_images(images_batch.numpy())
```

## 6- Individual Image Visualization

```python
# Display an individual image: 

plt.imshow(images_batch[3].numpy())
plt.axis('off')
plt.show()
```

## 7- Train a Simple Model Using the Processed Dataset

- I create a simple convolutional neural network (CNN)

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax') 
])
```

## I print a summary of the model's architecture

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model:

history = model.fit(ds2, epochs=10, 
)
```

## 8- Data Splitting

```python
# Calculate the cardinality of the dataset for splitting:

ds2_cardinality = tf.data.experimental.cardinality(ds2).numpy()

# Split the dataset into training, validation, and test sets:

train_size = int(0.7 * ds2_cardinality)
val_size = int(0.15 * ds2_cardinality)
test_size = ds2_cardinality - train_size - val_size

train_ds = ds2.take(train_size)
val_ds = ds2.skip(train_size).take(val_size)
test_ds = ds2.skip(train_size + val_size)
```

## 9- Transfer Learning

```python
# I use a pre-trained model (VGG16) as a base:

base_model = tf.keras.applications.VGG16(
    include_top=False, input_shape=(256, 256, 3), weights='imagenet'
)

for layer in base_model.layers:
    layer.trainable = False

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with the split datasets:

history = model.fit(train_ds, epochs=10, validation_data=val_ds)
```

## 10- Model Evaluation

```python
# I evaluate the model on the test set:

test_loss, test_accuracy = model.evaluate(test_ds)
print("Test accuracy:", test_accuracy)
```

## 11- Advanced Visualization (Grad-CAM)

```python
import numpy as np
import cv2

# I Define a function to compute Grad-CAM heatmap:

def compute_grad_cam(model, image, layer_name):
    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(layer_name).output, model.output])
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(image)
        loss = predictions[:, np.argmax(predictions[0])]
    
    output = conv_outputs[0]
    grads = tape.gradient(loss, conv_outputs)[0]
    guided_grads = tf.cast(output > 0, 'float32') * tf.cast(grads > 0, 'float32') * grads
    
    weights = tf.reduce_mean(guided_grads, axis=(0, 1))
    cam = np.dot(output, weights)
    cam = cv2.resize(cam.numpy(), (256, 256))
    cam = np.maximum(cam, 0)
    heatmap = cam / np.max(cam)
    return heatmap

# random image from the test dataset:

sample_image, _ = next(iter(test_ds))

# I choose the last convolutional layer's in the model:

conv_layer_name = 'conv2d_2'

# Grad-CAM heatmap for the chosen image:

heatmap = compute_grad_cam(model, sample_image, conv_layer_name)

# Superimpose the heatmap on the original image:

heatmap = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)
superimposed_img = cv2.addWeighted(cv2.cvtColor(sample_image[0].numpy(), cv2.COLOR_RGB2BGR), 0.5, heatmap, 0.5, 0)

# Display the original image, heatmap, and superimposed image:

plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(sample_image[0].numpy())
plt.title('Original Image')
plt.axis('off')
plt.subplot(132)
plt.imshow(heatmap)
plt.title('Grad-CAM Heatmap')
plt.axis('off')
plt.subplot(133)
plt.imshow(superimposed_img)
plt.title('Superimposed Image')
plt.axis('off')
plt.show()
```

## 12- Model Interpretability (SHAP Values)

```python
import shap

# sample image from the test set for interpretation:

interpretation_image, _ = next(iter(test_ds))

# I create a SHAP explainer using the model's prediction function:

explainer = shap.Explainer(model, test_ds.map(lambda x, y: x))

# I calculate SHAP values for the chosen image:

shap_values = explainer.shap_values(interpretation_image)

# Visualize the SHAP values:

shap.image_plot(shap_values, -interpretation_image.numpy())
```
