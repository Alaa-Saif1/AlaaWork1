```python
#Create a list of 1000 random integers between 1 and 100000, then calculate the Z-Score to check 
#for the outliers. Consider values Z-Score > 2 as outliers 

import numpy as np

np.random.seed(42)

d = np.random.randint(1,100000, size = 1000)

z_score = (d - np.mean(d)) / np.std(d)

outliers = d[abs(z_score) > 2]

outliers_d = pd.DataFrame({'Outliers' : outliers })

print(outliers_d)
```

```python
# Create a list of 10 random integers between 1 and 100.

np.random.seed(42)
d = np.random.randint(1,100, size = 10)
s = pd.Series(d)

# Standardize the numbers

mean = s.mean()
std_dev = s.std()
stand = (s - mean) / std_dev

print("standardized Data:")
print(stand)

# Normalize the numbers

nor = (s - s.min()) / (s.max() - s.min())

print("\nnormalized Data:")
print(nor)
```

```python
# Create a list of 100 random pair of integers (x,y) between 1 and 100. Draw scatter plot of the numbers

np.random.seed(42)
x = np.random.randint(1,100, size = 100)
y = np.random.randint(1,100, size = 100)

d = pd.DataFrame({'x':x, 'y':y})

d.plot(kind = 'scatter', x='x', y='y')
plt.show()
```
